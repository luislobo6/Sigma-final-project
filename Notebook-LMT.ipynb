# Transform

* From the pivot table the column cve_institucion only have the first value for each Bank, so we need to complete the empty fields with the same information and we use the "front fill method"
* Get the columns that represent the years and drop the empty ones

```python
# fill the empty spaces
data[["cve_institucion"]] = data[["cve_institucion"]].fillna(method="ffill")
data
```

```python
# Get the YearMonth columns to check only empty rows in this columns
cols = data.columns
cols = cols[2:-1]
cols
```

```python
data_clean = data.drop_duplicates(subset=cols, keep=False)
data_clean
```

```python
# After cleaning the dataframe we get a lighter dataframe
print(f"Initial: ${data.shape} - Now: ${data_clean.shape}")
```

```python
# Drop the rows where the cve_institucion is empty
data_clean = data_clean.dropna(subset=["cve_concepto"])
data_clean
```

```python
# check for null or na values
data_clean["cve_institucion"] = data_clean["cve_institucion"].astype(str)
data_clean["cve_institucion"].unique()
```

```python
# check for null or na values
data_clean["cve_concepto"].unique()
```

```python
# Filter data important for the project

#130000000000 = Cartera de crédito vigente
#130100000000 = Créditos comerciales
#131100000000 = Consumo
#131600000000 = Vivienda
#200000000000 = Pasivo
#210100000000 = Depósitos de exigibilidad inmediata
#211100000000 = Depósitos a plazo y títulos de crédito emitidos 
cves = [510000000000,610000000000,620000000000,670200000000,640000000000,
        670500000000,671100000000,110000000000,130100000000,
        130105000000,130106000000,130121000000,131100000000,131101000000,
        131199990000,131600000000,200000000000,210100000000,211100000000]

filter_cve = data_clean["cve_concepto"] == 100000000000
data_filtered = data_clean.loc[filter_cve]

for cve in cves:
    filter_cve = data_clean["cve_concepto"] == cve
    data_filtered = data_filtered.append(data_clean.loc[filter_cve])

```

```python
data_filtered.shape
```

```python
#get columns for the time
clean_cols = data_filtered.columns
clean_cols = clean_cols[2:-1]
clean_cols
```

```python
#rename the columns to add a prefix "value" so then we can make the wide_to_long
for column in clean_cols:
    data_filtered = data_filtered.rename(columns={
        column: f"value{column}" 
    })
```

```python
#check if it was done properly
data_filtered.columns
```

```python
# wide_to_long transformation so we have a vertical table with time series and values
data_filtered = pd.wide_to_long(data_filtered, stubnames='value', i=['cve_institucion', 'cve_concepto'], j='time')
```

```python
#reset index to have our 3 columns in dataframe
data_filtered = data_filtered.reset_index()
data_filtered
```

```python
del data_filtered["Grand Total"]
```

```python
# put the column time in time type data
data_filtered["time"] = pd.to_datetime(data_filtered["time"],format="%Y%m")
```

```python
data_filtered["value"] = data_filtered["value"].fillna(0)
```

```python
data_filtered
```

```python
# Save the file to continue cleaning in different sessions
data_filtered.to_csv("data/data_filtered.csv")
```
