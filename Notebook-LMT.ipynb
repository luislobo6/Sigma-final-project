# Transform

* From the pivot table the column cve_institucion only have the first value for each Bank, so we need to complete the empty fields with the same information and we use the "front fill method"
* Get the columns that represent the years and drop the empty ones

```python
# fill the empty spaces
data[["cve_institucion"]] = data[["cve_institucion"]].fillna(method="ffill")
data
```

```python
# Change column names to datetime format to use in the ML Facebook Prophet library and in the visualizations
for column in cols:
    data_clean = data_clean.rename(columns={
        column: datetime.strptime(str(column), '%Y%m') 
    })
```

```python
# Change column names to String to make them index and erase the .00 on each column
data_clean["cve_concepto"] = data_clean["cve_concepto"].astype(str).str.slice(stop=-2)
data_clean = data_clean.set_index(['cve_institucion', 'cve_concepto'])

data_clean
```

```python
# Transpose the matrix to use the data in the way the ML library needs it
data_clean_transpose = data_clean.T
data_clean_transpose
```

```python
# Join the two column indexes to have just one header or each column
data_clean_transpose.columns = ['_'.join(col) for col in data_clean_transpose.columns]
data_clean_transpose
```
